```python
sql_statement = """SELECT Authors, MyUnknownColumn as Author1, `MyUnknownColumn_[0]` as Author2, `ï»¿Accession Number`
              FROM LibBooks"""

table_df = pd.read_sql(
    sql_statement,
    con=engine
)
table_df
```

```python
# author insertion
import numpy as np

authors = table_df['Authors']
author1 = table_df['Author1']
author2 = table_df['Author2']
all_authors = np.concatenate((authors, author1, author2))
distinct_authors = np.unique(all_authors)

for author_name in distinct_authors:
    if author_name == '':
        continue
        
    query = "INSERT INTO authors (name) VALUES (%s)"
    cursor.execute(query, (author_name,))
```

```python
# book_author many-to-many table creation
for idx, row in table_df.iterrows():
    accession_no = row['ï»¿Accession Number']
    author = row['Authors']
        
    query = "INSERT INTO book_author (author_name, book_accession) VALUES (%s, %s)"
    cursor.execute(query, (author, accession_no))
    
    if row['Author1'] != '':
        author1 = row['Author1']
        cursor.execute(query, (author1, accession_no))
    
    if row['Author2'] != '':
        author2 = row['Author2']
        cursor.execute(query, (author2, accession_no))
```
